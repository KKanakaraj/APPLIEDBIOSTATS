# Load Data


```r
airline <- read.fwf("http://www.stat.ufl.edu/~winner/data/airline_costs.dat", 
widths=c(20,8,8,8,8,8,8,8,8,8,8,8),header=FALSE)
names(airline) <- c("airline","flight_length(mi)","plane_speed(mph)",
                    "daily_flight_time_plane(h)",
                    "population_served(1000s)",
                    "total_op_cost(cents per revenue ton-mile)", 
                    "revenue tons per Aircraft mile", 
                    "ton-mile load factor(proportion)",
                    "available capacity (tons per mile)", 
                    "total assets($100,000s)",
                    "investments and special funds($100,000s)", 
                    "adjusted assets($100,000s)")
```

```r
y <- airline[,12] # assign y to adjusted assets
x <- airline[,5]  # assign x to population served
linreg <- lm(y~x) # linear regression function
summ <- summary(linreg) # summary of parameters
```

```r
assets_population <- xyplot(airline[,12] ~ airline[,5], 
                            xlab=names(airline)[5],
                            ylab=names(airline)[12],
                               panel = function(x, y, ...) {
                                  panel.xyplot(x, y, ...)
                                  linreg = lm(y ~ x)
                                  summ = summary(linreg)
                                  r2 = lm1sum$adj.r.squared
                                  panel.abline(a = lm1$coefficients[1], 
                                            b = lm1$coefficients[2])
                                  panel.text(labels = 
                                    bquote(italic(R)^2 == 
                                    .(format(r2, digits = 3))),
                                         x = 4, y = 1000)
                            },                          
                            as.table = TRUE)
assets_population
```
